```python
music_sparse = csr_matrix(pivot_table)
```

```python
# Let's import the necessary libraries
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric="cosine", algorithm="brute")
model_knn.fit(music_sparse)
```

```python
def recommend_music(music_name):
    music_index = np.where(pivot_table.index == music_name)[0][0]
    distance, suggestion = model_knn.kneighbors(pivot_table.iloc[music_index,:].values.reshape(1,-1), n_neighbors=6)

    for i in range(len(suggestion)):
        music = pivot_table.index[suggestion[i]]
        for j in music:
            if j == music_name:
                print(f"You searched for {music_name} \n")
                print("The suggested songs are: \n")

            else:
                print(j)
```
