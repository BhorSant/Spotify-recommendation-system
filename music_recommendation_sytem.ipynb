```python
# Now removing the space like that
def remove_space(L):
    L1 = []
    for i in ast.literal_eval(L):
        L1.append(i['name'].replace(" ",""))
    return L1
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    T = []
    for i in text.split():
        T.append(ps.stem(i))
    return " ".join(T)
```
